```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = pd.read_csv('ecg.csv')
f
```

```python
plt.figure(figsize=(16, 8))
plt.plot(f['ecg'], linewidth = 1)
plt.show()
```

```python
n = len(f)
mu = f['ecg'].mean()
d = ((f['ecg'] - mu)**2).sum() / n
```

```python
r = []
for k in range(n):
	l = 1.0 / ((n - k) * d)
	sum = 0.0
	for j in range(n - k):
		sum = sum + (f['ecg'][j + k] - mu) * (f['ecg'][j] - mu)
	r.append(l * sum)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(r, linewidth = 1)
plt.show()
```
