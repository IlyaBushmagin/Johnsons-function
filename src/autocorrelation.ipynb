```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = pd.read_csv('ecg.csv')
f
```

```python
plt.figure(figsize=(16, 8))
plt.plot(f['ecg'], linewidth = 1)
plt.show()
```

```python
n = len(f)
mu = f['ecg'].mean()
d = ((f['ecg'] - mu)**2).sum() / n
```

```python
a = []
r = []
for k in range(n):
	l_a = 1.0 / (n - k)
	l_r = 1.0 / ((n - k) * d)
	sum_a = 0.0
	sum_r = 0.0
	for j in range(n - k):
		sum_a = sum_a + abs(f['ecg'][j + k] - f['ecg'][j])
		sum_r = sum_r + (f['ecg'][j + k] - mu) * (f['ecg'][j] - mu)
	a.append(l_a * sum_a)
	r.append(l_r * sum_r)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(a, linewidth = 1)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.plot(r, linewidth = 1)
plt.show()
```
